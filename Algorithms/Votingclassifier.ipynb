```python
import pandas as pd
# Load the dataset
file_path = 'training_data.csv'
data = pd.read_csv(file_path)
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# 编码目标变量
le = LabelEncoder()
data['increase_stock'] = le.fit_transform(data['increase_stock'])

# 将数据集分割为特征和目标
X = data.drop('increase_stock', axis=1)
y = data['increase_stock']

# 将数据分割为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# 设置单个模型
# Correcting the logistic regression instantiation and setting up models for soft voting
lr = LogisticRegression(max_iter=1600)
lda = LinearDiscriminantAnalysis()
qda = QuadraticDiscriminantAnalysis()
knn = KNeighborsClassifier(n_neighbors=2)
random_forest = RandomForestClassifier()

# Combining models into a soft voting classifier
soft_voting_classifier = VotingClassifier(estimators=[
    ('lr', lr),
    ('lda', lda),
    ('qda', qda),
    ('knn', knn),
    ('random_forest', random_forest)
], voting='soft')

# Fitting the soft voting classifier
soft_voting_classifier.fit(X_train, y_train)

# Making predictions and evaluating the model
y_pred_soft = soft_voting_classifier.predict(X_test)
accuracy_soft = accuracy_score(y_test, y_pred_soft)

accuracy_soft

```
